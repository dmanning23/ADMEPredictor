# 01 – Data Exploration

Explore the ADMET-AI dataset: distributions, missing values, molecular property ranges.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# TODO: update path after downloading dataset
df = pd.read_csv('../data/admet_ai.csv')
df.head()
```

```python
df.describe()
```

```python
# Distribution of target properties
target_cols = ['solubility', 'permeability', 'cyp3a4_inhibitor', 'herg_blocker']
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for ax, col in zip(axes.flat, target_cols):
    df[col].dropna().hist(ax=ax, bins=40)
    ax.set_title(col)
plt.tight_layout()
plt.show()
```
